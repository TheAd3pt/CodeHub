#### f) Same as above (7.e.), depending on whether or not we have low/high quantity (variable we created for 6.f.).

```python
# Count how many 'High' observations we have are in 'BinCuantity' column (Quantity >= 10)
counthigh = RawConstruction_Data[['Quantity','TotalCost','BinCuantity']].groupby('BinCuantity').count()['Quantity'][0]

# Count how many 'Low' observations we have are in 'BinCuantity' column (Quantity <= 10)
countlow = RawConstruction_Data[['Quantity','TotalCost','BinCuantity']].groupby('BinCuantity').count()['Quantity'][1]
```

```python
# Filter DataFrame with 'High' quantity by sorting and counting until the observation refers to 'High'. Drop the column of
# 'BinCuantity'.
high = RawConstruction_Data[['Quantity','TotalCost','BinCuantity']].sort_values(by='BinCuantity')[0:counthigh].drop('BinCuantity',axis=1)

# Filter DataFrame with 'Low' quantity by sorting and counting from the observation refers to 'Low' until the end. 
# Drop the column of 'BinCuantity'.
low = RawConstruction_Data[['Quantity','TotalCost','BinCuantity']].sort_values(by='BinCuantity')[counthigh:countlow].drop('BinCuantity',axis=1)
```

```python
# Extract outliers  for high category
high_quantity = high['Quantity'].as_matrix()
high_quantity = st.trimboth(high_quantity,0.03)
high_TotalCost = high['TotalCost'].as_matrix()
high_TotalCost = st.trimboth(high_TotalCost,0.03)
```

```python
df_high = pd.DataFrame(high_quantity,high_TotalCost).reset_index().rename(columns = {'index':'TotalCost', 0:'Quantity'})
```

```python
plt.figure(figsize=(7,7))
sns.regplot(x='Quantity',y='TotalCost',data=df_high[['Quantity','TotalCost']],color = 'green')
plt.show()
```

```python
# Extract outliers for low category
low_quantity = low['Quantity'].as_matrix()
low_quantity = st.trimboth(low_quantity,0.02)
low_TotalCost = low['TotalCost'].as_matrix()
low_TotalCost = st.trimboth(low_TotalCost,0.02)
```

```python
df_low = pd.DataFrame(low_quantity,low_TotalCost).reset_index().rename(columns = {'index':'TotalCost', 0:'Quantity'})
```

```python
plt.figure(figsize=(7,7))
sns.regplot(x='Quantity',y='TotalCost',data=df_low[['Quantity','TotalCost']],color = 'purple')
plt.ylim((0,300))
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax1 = fig.add_subplot(111)
sns.regplot(x='Quantity',y='TotalCost',data= df_high[['Quantity','TotalCost']],ax=ax1,color = 'green')
sns.regplot(x='Quantity',y='TotalCost',data= df_low[['Quantity','TotalCost']],ax=ax1,color = 'red')
plt.ylim((0,600))
plt.xlim((0,200))
plt.show()
```
